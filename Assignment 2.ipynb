## Question 1

Credit card numbers follow certain patterns. A credit card number must have between 13 and 16 digits. It must start with:<br>
4 for Visa cards<br>
5 for Master cards<br>
37 for American Express cards<br>
6 for Discover cards<br>

In 1954, Hans Luhn of IBM proposed an algorithm for validating credit card numbers. The algorithm is useful to determine if a card number is entered correctly or if a credit card is scanned correctly by a scanner. Almost all credit card numbers are generated following this validity check, commonly known as the Luhn check or the Mod 10 check, which can be described as follows (for illustration, consider the card number 4388576018402626):<br>

1. Double every second digit from right to left. If doubling of a digit results in a two-digit number, add up the two digits to get a single-digit number.<br>
2 * 2 = 4<br>
2 * 2 = 4<br>
4 * 2 = 8<br>
1 * 2 = 2<br>
6 * 2 = 12 (1 + 2 = 3)<br>
5 * 2 = 10 (1 + 0 = 1)<br>
8 * 2 = 16 (1 + 6 = 7)<br>
4 * 2 = 8

2. Now add all single-digit numbers from Step 1. <br>
4 + 4 + 8 + 2 + 3 + 1 + 7 + 8 = 37<br>

3. Add all digits in the odd places from right to left in the card number.<br>
   6 + 6 + 0 + 8 + 0 + 7 + 8 + 3 = 38<br>

4. Sum the results from Step 2 and Step 3.<br>
37 + 38 = 75<br>

5. If the result from Step 4 is divisible by 10, the card number is valid; otherwise, it is invalid. For example, the number 4388576018402626 is invalid, but the number 4388576018410707 is valid.<br>

Write a program that prompts the user to enter a credit card number as a long integer. Display whether the number is valid or invalid. <br>

Here are sample runs of the program:<br>

Sample 1:<br>
Enter a credit card number as a long integer: 4246345689049834<br>
4246345689049834 is invalid <br>

Sample 2:<br>
Enter a credit card number as a long integer: 4388576018410707<br>
4388576018410707 is valid <br>

```python
def card_num_input():
    card_num = input("Please enter a 10-digit number: ")
    if len(card_num) == 10 and card_num.isdigit():
        return card_num
    else:
        print("Invalid input. Please enter exactly 10 digits.")
        return None

def num_check(card_num):
    reverse_digits = card_num[::-1]
    total = 0
    for i, digit in enumerate(reverse_digits):
        n = int(digit)
        if (i + 1) % 2 == 0:  # double every second digit
            n *= 2
            if n > 9:
                n -= 9
        total += n
    return total

def is_valid(total, card_num):
    if total % 10 == 0:
        print(f"{card_num} is valid")
    else:
        print(f"{card_num} is invalid")

card_num = card_num_input()
if card_num is not None:
    total = num_check(card_num)
    is_valid(total, card_num)
```

## Question 2
Design a class named Complex for representing complex numbers and the methods __add__, __sub__, __mul__, __truediv__, __abs__ for performing complex-number operations, and override __str__ method for returning a string representation for a complex number (Note that all these methods have double undescores before and after the ir names). The __str__ method returns (a + bi) as a string. If b is 0, it simply returns a. <br>

Provide a constructor Complex(a, b) to create a complex number a + bi with default value for a and b as 0. Also provide the getRealPart() and getImaginaryPart() methods for returning the real and imaginary part of the complex number, respectively. <br>
Write a test program that prompts the user to enter two complex numbers and display the result of their addition, subtraction, multiplication, and division. Here is a sample run: <br>

<Output> <br>
Enter the first complex number: 3.5, 5.5 <br>
Enter the second complex number: -3.5, 1 <br>
(3.5 + 5.5i) + (-3.5 + 1.0i) = 0.0 + 6.5i <br>
(3.5 + 5.5i) - (-3.5 + 1.0i) = 7.0 + 4.5i <br>
(3.5 + 5.5i) * (-3.5 + 1.0i) = -17.75 + -15.75i <br>
(3.5 + 5.5i) / (-3.5 + 1.0i) = -0.5094 + -1.7i <br>
|(3.5 + 5.5i)| = 6.519202405202649 <br>
<End Output>


```python
class Complex:
    def __init__(self, a=0, b=0):
        self.a = a
        self.b = b

    def get_real_part(self):
        return self.a

    def get_imaginary_part(self):
        return self.b

    def __add__(self, other):
        return Complex(self.a + other.a, self.b + other.b)

    def __sub__(self, other):
        return Complex(self.a - other.a, self.b - other.b)

    def __mul__(self, other):
        real = self.a * other.a - self.b * other.b
        imag = self.a * other.b + self.b * other.a
        return Complex(real, imag)

    def __truediv__(self, other):
        if other.a == 0 and other.b == 0:
            raise ZeroDivisionError("division by zero")
        denominator = other.a ** 2 + other.b ** 2
        real = (self.a * other.a + self.b * other.b) / denominator
        imag = (self.b * other.a - self.a * other.b) / denominator
        return Complex(real, imag)

    def __abs__(self):
        return (self.a ** 2 + self.b ** 2) ** 0.5

    def __str__(self):
        if self.b == 0:
            return str(self.a)
        if self.b > 0:
            return f"({self.a} + {self.b}i)"
        return f"({self.a} - {-self.b}i)"

def main():
    def get_complex_input(prompt):
        while True:
            try:
                a, b = map(float, input(prompt).split(", "))
                return a, b
            except ValueError:
                print("Invalid input. Please enter two numbers separated by a comma (e.g., 3, 4).")

    a1, b1 = get_complex_input("Enter the first complex number (format: a, b): ")
    a2, b2 = get_complex_input("Enter the second complex number (format: a, b): ")

    c1 = Complex(a1, b1)
    c2 = Complex(a2, b2)

    try:
        print(f"{c1} + {c2} = {c1 + c2}")
        print(f"{c1} - {c2} = {c1 - c2}")
        print(f"{c1} * {c2} = {c1 * c2}")
        print(f"{c1} / {c2} = {c1 / c2}")
    except ZeroDivisionError as e:
        print(e)

    print(f"|{c1}| = {abs(c1)}")
    print(f"|{c2}| = {abs(c2)}")

if __name__ == "__main__":
    main()
```

## Question 3

Write a program that removes all the occurrences of a specified string from a text file. Your program should prompt the user to enter a file and a string to be removed (you do not need to attach a test case for this and also next question, and the TA will test the code with her own file). Here is a sample run:

<Output><br>

Enter the string to be removed: morning<br>
Done<br>
<End Output>

```python
def remove_string_from_file():
    # Prompt user for file name and string to remove
    filename = input("Enter a filename: ")
    string_to_remove = input("Enter the string to be removed: ")

    try:
        # Read the contents of the file
        with open(filename, 'r') as file:
            contents = file.read()

        # Remove the string
        contents = contents.replace(string_to_remove, '')

        # Write the updated contents back to the file
        with open(filename, 'w') as file:
            file.write(contents)

        print("Done")

    except FileNotFoundError:
        print("Error: The file does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
remove_string_from_file()
```

## Question 4
Write the following function that tests whether a two-dimensional list has four consecutive numbers of the same value, either horizontally, vertically, or diagonally.<br>

def isConsecutiveFour(values): <br>

Write a test program that prompts the user to enter the number of rows and columns of a two-dimensional list and then the values in the list and displays True if the list contains four consecutive numbers with the same value. Otherwise, display False. Here are some examples of the true cases:<br>
    

```python
def isConsecutiveFour(values):
    rows = len(values)
    cols = len(values[0]) if values else 0

    # Check horizontally
    for r in range(rows):
        for c in range(cols - 3):
            if values[r][c] == values[r][c+1] == values[r][c+2] == values[r][c+3]:
                return True

    # Check vertically
    for c in range(cols):
        for r in range(rows - 3):
            if values[r][c] == values[r+1][c] == values[r+2][c] == values[r+3][c]:
                return True

    # Check diagonal (top-left to bottom-right)
    for r in range(rows - 3):
        for c in range(cols - 3):
            if values[r][c] == values[r+1][c+1] == values[r+2][c+2] == values[r+3][c+3]:
                return True

    # Check diagonal (top-right to bottom-left)
    for r in range(rows - 3):
        for c in range(3, cols):
            if values[r][c] == values[r+1][c-1] == values[r+2][c-2] == values[r+3][c-3]:
                return True

    return False
```

## Question 5

Design a class named Triangle that extends GeometricObject. The class contains: <br>
•	Three float data fields named side1, side2, and side3 to denote three sides of the triangle.<br>
•	A constructor that creates a triangle with the specified side1, side2, and side3 with default values 1.0.<br>
•	The accessor methods for all three data fields. <br>
•	A method named getArea() that returns the area of this triangle.<br>
•	A method named getPerimeter() that returns the perimeter of this triangle.<br>
•	A method named __str__() that returns a string description for the triangle. <br>

For the formula to compute the area of a triangle, see Exercise 2.21. The __str__() method is implemented as follows:<br>

return "Triangle: side1 = " + str(side1) + " side2 = " + <br>
    str(side2) + " side3 = " + str(side3)<br>

Implement the class. Write a test program that creates a Triangle object with sides 1, 1.5, 1, color yellow and filled true, and displays the area, perimeter, color, and whether filled or not.


```python
class GeometricObject:
    def __init__(self, color="white", filled=False):
        self._color = color
        self._filled = filled

    def get_color(self):
        return self._color

    def set_color(self, color):
        self._color = color

    def is_filled(self):
        return self._filled

    def set_filled(self, filled):
        self._filled = filled

    def __str__(self):
        return f"GeometricObject: color = {self._color}, filled = {self._filled}"

class Triangle(GeometricObject):
    def __init__(self, side1=1.0, side2=1.0, side3=1.0, color="white", filled=False):
        super().__init__(color, filled)
        self.set_sides(side1, side2, side3)

    def set_sides(self, side1, side2, side3):
        if not self.is_valid_triangle(side1, side2, side3):
            raise ValueError("The provided sides do not form a valid triangle")
        self._side1 = float(side1)
        self._side2 = float(side2)
        self._side3 = float(side3)

    def get_side1(self):
        return self._side1

    def set_side1(self, side1):
        self.set_sides(side1, self._side2, self._side3)

    def get_side2(self):
        return self._side2

    def set_side2(self, side2):
        self.set_sides(self._side1, side2, self._side3)

    def get_side3(self):
        return self._side3

    def set_side3(self, side3):
        self.set_sides(self._side1, self._side2, side3)

    def get_area(self):
        s = self.get_perimeter() / 2
        return (s * (s - self._side1) * (s - self._side2) * (s - self._side3)) ** 0.5

    def get_perimeter(self):
        return self._side1 + self._side2 + self._side3

    def is_valid_triangle(self, side1, side2, side3):
        side1, side2, side3 = float(side1), float(side2), float(side3)
        return (side1 + side2 > side3 and
                side1 + side3 > side2 and
                side2 + side3 > side1)

    def __str__(self):
        return (f"Triangle: side1 = {self._side1}, "
                f"side2 = {self._side2}, side3 = {self._side3}")

# Test Case
def main():
    try:
        side1 = float(input("Enter side1 of the triangle: "))
        side2 = float(input("Enter side2 of the triangle: "))
        side3 = float(input("Enter side3 of the triangle: "))
        color = input("Enter the color of the triangle: ")
        filled = input("Is the triangle filled? (yes/no): ").lower() == 'yes'

        triangle = Triangle(side1, side2, side3, color, filled)
        print(triangle)
        print("Area:", triangle.get_area())
        print("Perimeter:", triangle.get_perimeter())
        print("Color:", triangle.get_color())
        print("Filled:", triangle.is_filled())
    except ValueError as e:
        print(e)

if __name__ == "__main__":
    main()
```

## Question 6

(The Account class) Design a class named Account that contains: <br>
•	A private int data field named id for the account (default 0).<br>
•	A private double data field named balance for the account (default 0).<br>
•	A private double data field named annualInterestRate that stores the current interest rate (default 0). Assume all accounts have the same interest rate.<br>
•	A private Date data field named dateCreated that stores the date when the account was created.<br>
•	A no-arg constructor that creates a default account.<br>
•	A constructor that creates an account with the specified id and initial balance.<br>
•	The accessor and mutator methods for id, balance, and annualInterestRate. <br>
•	The accessor method for dateCreated. <br>
•	A method named getMonthlyInterestRate() that returns the monthly interest rate.<br>
•	A method named withdraw that withdraws a specified amount from the account.<br>
•	A method named deposit that deposits a specified amount to the account.<br>

Write a test program that creates an Account object with an account ID of 1122, a balance of 20,000 CAD, and an annual interest rate of 4.5 percent. Use the withdraw method to withdraw 2,500 CAD, use the deposit method to deposit 3,000 CAD, and print the balance, the monthly interest, and the date when this account was created.

```python
from datetime import datetime

class Account:
    def __init__(self, id=0, balance=0.0):
        self._id = id
        self._balance = balance
        self._annual_interest_rate = 0.0
        self._date_created = datetime.now()

    def get_id(self):
        return self._id

    def get_balance(self):
        return self._balance

    def get_annual_interest_rate(self):
        return self._annual_interest_rate

    def get_date_created(self):
        return self._date_created

    def set_id(self, id):
        if id < 0:
            print("ID cannot be negative")
        else:
            self._id = id

    def set_balance(self, balance):
        if balance < 0:
            print("Balance cannot be negative")
        else:
            self._balance = balance

    def set_annual_interest_rate(self, annual_interest_rate):
        if annual_interest_rate < 0:
            print("Annual interest rate cannot be negative")
        else:
            self._annual_interest_rate = annual_interest_rate

    def get_monthly_interest_rate(self):
        return self._annual_interest_rate / 12

    def withdraw(self, amount):
        try:
            if amount < 0:
                raise ValueError("Withdraw amount cannot be negative")
            elif amount > self._balance:
                raise ValueError("Insufficient balance")
            else:
                self._balance -= amount
        except ValueError as e:
            print(e)

    def deposit(self, amount):
        try:
            if amount < 0:
                raise ValueError("Deposit amount cannot be negative")
            else:
                self._balance += amount
        except ValueError as e:
            print(e)

# Test Case
def main():
    id = int(input("Enter account ID: "))
    balance = float(input("Enter initial balance (CAD): "))
    annualInterestRate = float(input("Enter annual interest rate (%): "))

    account = Account(id, balance)
    account.set_annual_interest_rate(annualInterestRate)

    while True:
        action = input("Enter 'w' to withdraw, 'd' to deposit, or 'q' to quit: ").lower()
        if action == 'q':
            break
        elif action == 'w':
            amount = float(input("Enter amount to withdraw (CAD): "))
            account.withdraw(amount)
        elif action == 'd':
            amount = float(input("Enter amount to deposit (CAD): "))
            account.deposit(amount)
        else:
            print("Invalid action. Please enter 'w', 'd', or 'q'.")

    print(f"Balance: {account.get_balance()} CAD")
    print(f"Monthly Interest: {account.get_monthly_interest_rate()}%")
    print(f"Date Created: {account.get_date_created()}")

if __name__ == "__main__":
    main()
```
