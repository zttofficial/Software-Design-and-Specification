````python
# Write the code for MediaPlayer Interface here

from abc import ABC, abstractmethod

class MediaPlayer(ABC):
    @abstractmethod
    def play(self, file_name):
        pass
````

````python
#Write the code for Concrete Media Players here

# MP3Player Class
class MP3Player(MediaPlayer):
    def play(self, file_name):
        print(f"Playing MP3 file: {file_name}")

# MP4Player Class
class MP4Player(MediaPlayer):
    def play(self, file_name):
        print(f"Playing MP4 file: {file_name}")

# AVIPlayer Class
class AVIPlayer(MediaPlayer):
    def play(self, file_name):
        print(f"Playing AVI file: {file_name}")
````

````python
#Write the code for MediaAdapter Class here

class MediaAdapter(MediaPlayer):
    def __init__(self, media_type):
        if media_type == 'mp3':
            self.media_player = MP3Player()
        elif media_type == 'mp4':
            self.media_player = MP4Player()
        elif media_type == 'avi':
            self.media_player = AVIPlayer()

    def play(self, file_name):
        self.media_player.play(file_name)
````

````python
#Write the code for Write the code here

import pickle

class MediaLibrary:
    def __init__(self):
        self.media_files = []

    def add_media(self, file_name, media_type):
        self.media_files.append((file_name, media_type))

    def view_media(self):
        return self.media_files

    def save_library(self, file_path):
        with open(file_path, 'wb') as file:
            pickle.dump(self.media_files, file)
        print(f"Media library saved to {file_path}")

    def load_library(self, file_path):
        with open(file_path, 'rb') as file:
            self.media_files = pickle.load(file)
        print(f"Media library loaded from {file_path}")
````

````python
#Write the main function and test the code here

def main():
    library = MediaLibrary()
    while True:
        print("\nMedia Player\n")
        print("1. Add Media File")
        print("2. Play Media File")
        print("3. View Media Library")
        print("4. Save Media Library")
        print("5. Load Media Library")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            file_name = input("Enter media file name: ")
            media_type = input("Enter media type (mp3, mp4, avi): ").lower()
            library.add_media(file_name, media_type)
        elif choice == '2':
            file_name = input("Enter media file name to play: ")
            media_type = input("Enter media type (mp3, mp4, avi): ").lower()
            adapter = MediaAdapter(media_type)
            adapter.play(file_name)
        elif choice == '3':
            media_files = library.view_media()
            for file_name, media_type in media_files:
                print(f"{file_name} ({media_type})")
        elif choice == '4':
            file_path = "media_library.pkl"
            library.save_library(file_path)
        elif choice == '5':
            file_path = "media_library.pkl"
            library.load_library(file_path)
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
````
